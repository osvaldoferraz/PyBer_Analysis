```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create a plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, label='Boston', color='green', marker='*', linewidth=2)
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_ylim(0,45)
ax.grid()
ax.set_title('PyBer Fare by Month')
ax.legend()
```

```python
# SKILL DRILL
plt.scatter(y_axis, x_axis, color='red', label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.gca().invert_yaxis()
plt.legend()
plt.xlim(0, 50)
```

```python
# Bubble chart

# make y_axis plots 3x times bigger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)  
  
    
    
```

```python
# Conditionals using 'list comprehension'
plt.scatter(x_axis, y_axis, s = [i *3 for i in y_axis])
```

```python
# Create scatter using ax.scatter SKILL DRILL
fig, ax = plt.subplots()
ax.scatter(
    y_axis, x_axis, 
    c = 'skyblue', 
    s = [i * 5 for i in y_axis], 
    alpha=0.8, 
    edgecolor = 'black', 
    linewidth=2,
    label = 'Boston'
)
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
ax.legend()
ax = plt.gca().invert_yaxis()



```

```python
# create a pie chart with plt.pie
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# explode one slice of the pie
plt.subplots(figsize=(10,10))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assing 12 colors, one for each month
colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen', 'greenyellow', 'yellow', 'orange', 'gold', 'indianred', 'tomato', 'mistyrose']
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(10,10))
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')
plt.show()
```
